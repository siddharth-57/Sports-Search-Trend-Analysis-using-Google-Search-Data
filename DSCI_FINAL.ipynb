```python
from pytrends.request import TrendReq
import plotly.graph_objs as go
import plotly.offline as pyo

# set Google Trends API parameters
pytrends = TrendReq(hl='en-US', tz=360)

# define keywords to get data
kw_list = ["Cricket","Basketball", "Football"]

# build payload
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='', gprop='')

# fetch interest over time data
interest_over_time_df = pytrends.interest_over_time()

# fetch related queries and topics data
related_queries_dict = {}
related_topics_dict = {}
for kw in kw_list:
    related_queries_dict[kw] = pytrends.related_queries()[kw]['top']
    if kw in pytrends.related_topics():
        related_topics_dict[kw] = pytrends.related_topics()[kw]['title'].tolist()

# plot interest over time using Plotly
data = []
for kw in kw_list:
    trace = go.Scatter(
        x=interest_over_time_df.index,
        y=interest_over_time_df[kw],
        name=kw
    )
    data.append(trace)

layout = go.Layout(
    title='Interest Over Time',
    xaxis=dict(
        title='Date'
    ),
    yaxis=dict(
        title='Interest Score'
    )
)

fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)

# print related queries and topics for each keyword
for kw in kw_list:
    print(f"Related queries for {kw}:")
    print(related_queries_dict[kw])
    if kw in related_topics_dict:
        print(f"\nRelated topics for {kw}:")
        print(related_topics_dict[kw])
    print("\n")

```

```python
# create a new column 'airlines' with words with @
data['airlines'] = data['text'].apply(lambda x: ' '.join(re.findall(r'\@\w+', x)))

# remove words with @ from the original column and copy to processed column
data['processed'] = data['text'].apply(lambda x: re.sub(r'\@\w+', '', x))

data.head()
data.tail()
```

```python
# Function to preprocess Reviews data
def preprocess_Reviews_data(data,name):

    #1. removing quotes and ':'
    data[name]=data[name].apply(lambda x:re.sub(r"([\"“”:])", "",x))

    #2. (i) separate clitics into component words
    data[name]=data[name].apply(lambda x:re.sub(r"'m", " am", x))
    data[name]=data[name].apply(lambda x:re.sub(r"'d", " would", x))
    data[name]=data[name].apply(lambda x:re.sub(r"can't", "can not", x))
    data[name]=data[name].apply(lambda x:re.sub(r"n't", " not", x))
    data[name]=data[name].apply(lambda x:re.sub(r"'ll", " will", x))
    data[name]=data[name].apply(lambda x:re.sub(r"'ve", " have", x))
    data[name]=data[name].apply(lambda x:re.sub(r"'re", " are", x))

    # (ii) convert word-ending apostrophes to apostrophe s
    data[name]=data[name].apply(lambda x:re.sub(r"'s ", " is ", x))
    data[name]=data[name].apply(lambda x:re.sub(r"s' ", "s is ", x))

    # 3. separate hashtags and user references
    data[name]=data[name].apply(lambda x:re.sub(r"#(\w+)", r"# \1", x))
    data[name]=data[name].apply(lambda x:re.sub(r"@(\w+)", r"@ \1", x))

    #4. emojis to words


    # 5. merge words with spaces
    data[name]=data[name].apply(lambda x:re.sub(r"\b([A-Za-z]\.)\s+([A-Za-z]\.)\s+([A-Za-z]\.)\b", r"\1\2\3", x))
    data[name]=data[name].apply(lambda x:re.sub(r"\b([A-Za-z]\.)\s+([A-Za-z]\.)\b", r"\1\2", x))

    #6. removing "....." and replacing with a space
    data[name]=data[name].apply(lambda x:re.sub(r"\.{2,}", " ", x))

    # Code to remove the links from the text
    data[name]=data[name].apply(lambda x:re.sub(r"http\S+", "", x))

    # Code to remove the Special characters from the text
    data[name]=data[name].apply(lambda x:' '.join(re.findall(r'\w+', x)))

    # Remove the twitter handlers
    data[name]=data[name].apply(lambda x:re.sub('@[^\s]+','',x))

    # Proprocessing the data
    data[name]=data[name].str.lower()


    # # Code to remove the Hashtags from the text
    # data[name]=data[name].apply(lambda x:re.sub(r'\B#\S+','',x))





    # # Code to substitute the multiple spaces with single spaces
    # data[name]=data[name].apply(lambda x:re.sub(r'\s+', ' ', x, flags=re.I))
    # # Code to remove all the single characters in the text
    # data[name]=data[name].apply(lambda x:re.sub(r'\s+[a-zA-Z]\s+', '', x))




preprocess_Reviews_data(data,'processed')
data.to_csv('data_processed.csv', index=False)
data.head()
data.tail()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# replace '<1' values with 0
def clean_data(df):
    df = df.replace('<1', 0)
    return df

# read input CSV files and plot line graph for each file
def plot_data(files):
    for file in files:
        # read CSV file
        df = pd.read_csv(file, skiprows=1)

        # clean data
        df_cleaned = clean_data(df)

        # plot line graph
        plt.plot(df_cleaned['Month'], df_cleaned.drop('Month', axis=1))
        plt.legend(df_cleaned.drop('Month', axis=1).columns)
        plt.xlabel('Month')
        plt.ylabel('Number of Searches')
        plt.title(file[:-4])
        plt.show()

# specify input CSV files
files = ['file1.csv', 'file2.csv', 'file3.csv', 'file4.csv', 'file5.csv']

# plot line graph for each input file
plot_data(files)
```
